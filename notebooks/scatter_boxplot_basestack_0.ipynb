### K-mean clustering

```python
eigv_id = 1
eigv_sele = plot_agent.p_agent.s_agent.get_eigenvector_by_id(eigv_id)
```

```python
df_top5_percent.shape
```

```python
n_selection = df_top5_percent.shape[0]
eigv_mat = np.zeros((n_selection, plot_agent.p_agent.s_agent.n_node))
for eigv_id in range(1, n_selection+1):
    eigv_mat[eigv_id-1,:] = np.abs(plot_agent.p_agent.s_agent.get_eigenvector_by_id(eigv_id))
```

```python
n_clusters = 2
kmeans = KMeans(n_clusters=n_clusters, random_state=0).fit(eigv_mat)
```

```python
kmeans.labels_
```

```python
df_top5_percent = df_top5_percent.assign(kmeans_label=pd.Series(kmeans.labels_).values)

#df_top5_percent['kmeans_label'] = kmeans.labels_[:]
df_top5_percent
```

### Combine Plot

```python
fig = plt.figure(constrained_layout=True, figsize=(8,8))

gs = fig.add_gridspec(12, 1)
ax1 = fig.add_subplot(gs[0:2, 0])
ax2 = fig.add_subplot(gs[2:7, 0])
ax3 = fig.add_subplot(gs[7:, 0])

ax1.boxplot([df['Mean-Mode-lambda']], vert=False, widths=4, whis=(5,95))
ax1.set_axis_off()
ax1.set_title('A-tract-Strand1', fontsize=16)

ax2.scatter(df['Mean-Mode-lambda'], df['Mean-r-alpha'], s=6, color='grey')
ax2.scatter(df_top5_percent['Mean-Mode-lambda'], df_top5_percent['Mean-r-alpha'], s=6, color='blue')
ax2.set_xlabel(r'$\lambda_{\alpha}$', fontsize=14)
ax2.set_ylabel(r'$\left< r_{\alpha}\right>$', fontsize=14)
xlim_ax2 = ax2.get_xlim()
ax1.set_xlim(xlim_ax2)


offset_x = 0.05
offset_y = 0.001
for kmean_label in range(n_clusters):
    mask = df_top5_percent['kmeans_label'] == kmean_label
    df_sele = df_top5_percent[mask]
    ax3.scatter(df_sele['Mean-Mode-lambda'], df_sele['Mean-r-alpha'], s=10)
    
    text_x_list = df_sele['Mean-Mode-lambda'].tolist()
    text_y_list = df_sele['Mean-r-alpha'].tolist()
    text_list = df_sele['Mode-ID'].tolist()
    for text, x, y in zip(text_list, text_x_list, text_y_list):
            ax3.text(x+offset_x, y+offset_y, text)
    
ax3.set_xlabel(r'$\lambda_{\alpha}$', fontsize=14)
ax3.set_ylabel(r'$\left< r_{\alpha}\right>$', fontsize=14)

plt.show()
```
