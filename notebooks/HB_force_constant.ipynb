```python
import matplotlib.pyplot as plt
from enmspring.spring import Spring
from enmspring.k_b0_util import get_df_by_filter_bp
from enmspring.hb_util import HBAgent
```

### Part 0: Initialization

```python
rootfolder = '/home/yizaochen/codes/dna_rna/fluctmatch_sequence'
host = 'ctct_21mer'
n_bp = 21
h_agent = HBAgent(host, rootfolder, n_bp)
```

```python
h_agent.initialize_basepair()
```

### Part 1: G-tract and GCGC-21mer

```python
bptype = 'GC'
xticks = range(1, 22)
ylim = (-0.5, 10)

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(14,6))

typename = 'type1'
resids, klist = h_agent.get_resid_klist_by_type(bptype, typename)
ax.plot(resids, klist, '-o', label=typename)

typename = 'type2'
resids, klist = h_agent.get_resid_klist_by_type(bptype, typename)
ax.plot(resids, klist, '-o', label=typename)

typename = 'type3'
resids, klist = h_agent.get_resid_klist_by_type(bptype, typename)
ax.plot(resids, klist, '-o', label=typename)

ax.set_title(host, fontsize=14)
ax.legend(fontsize=14)
ax.set_xticks(xticks)
ax.axvline(4, color='red', alpha=0.5)
ax.axvline(18, color='red', alpha=0.5)
ax.tick_params(axis='both', labelsize=14)
ax.set_xlabel('Resid', fontsize=14)
ax.set_ylabel('k (kcal/mol/$^2$)', fontsize=14)
ax.set_ylim(ylim)

plt.tight_layout()
plt.savefig(f'{host}_hb_gc.png', dpi=200)
plt.show()
```
