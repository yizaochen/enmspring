### Additional Part 1: Single plot for a specific host

```python
host = 'gcgc_21mer'
n_bp = 21
h_agent = HBAgent(host, rootfolder, n_bp)
h_agent.initialize_basepair()
```

```python
xticks = range(1, 22)
ylim = (-0.5, 10)

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(14,6))

typename = 'type1'
resids, klist = h_agent.get_resid_klist_all(typename)
ax.plot(resids, klist, '-o', label=typename)

typename = 'type2'
resids, klist = h_agent.get_resid_klist_all(typename)
ax.plot(resids, klist, '-o', label=typename)

typename = 'type3'
resids, klist = h_agent.get_resid_klist_all(typename)
ax.plot(resids, klist, '-o', label=typename)

ax.set_title(host, fontsize=14)
ax.legend(fontsize=14)
ax.set_xticks(xticks)
ax.axvline(4, color='red', alpha=0.5)
ax.axvline(18, color='red', alpha=0.5)
ax.tick_params(axis='both', labelsize=14)
ax.set_xlabel('Resid', fontsize=14)
ax.set_ylabel('k (kcal/mol/$^2$)', fontsize=14)
ax.set_ylim(ylim)

plt.tight_layout()
#plt.savefig(f'{host}_hb_gc.png', dpi=200)
plt.show()
```
