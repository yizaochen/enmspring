### Part : Four systems eigenvector plot

```python
hosts = ['a_tract_21mer', 'tat_21mer', 'g_tract_21mer', 'gcgc_21mer']
interval_time = 500
d_agent = dict()
for host in hosts:
    d_agent[host] = StackMeanModeAgent(host, rootfolder, interval_time)
    d_agent[host].load_mean_mode_laplacian_from_npy()   
    d_agent[host].eigen_decompose()    
```

```python
fig, ax = plt.subplots(figsize=(8,4))

for host in hosts:
    eigenvalues = d_agent[host].w
    xarray = np.array(range(len(eigenvalues)))
    ax.plot(xarray, eigenvalues, label=host)
ax.legend()
ax.set_xlabel("Mode Index, i", fontsize=14)
ax.set_ylabel("$\lambda_i$", fontsize=14)
ax.set_title("Mean-mode eigenvalues for base-stacking", fontsize=16)

plt.tight_layout()
plt.show()
```
