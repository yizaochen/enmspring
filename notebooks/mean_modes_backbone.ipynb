### Part 4: Four systems prepare all required data

```python
hosts = ['a_tract_21mer', 'tat_21mer', 'g_tract_21mer', 'gcgc_21mer']
interval_time = 500
d_agent = dict()
for host in hosts:
    d_agent[host] = BackboneMeanModeAgent(host, rootfolder, interval_time)
    d_agent[host].load_mean_mode_laplacian_from_npy()   
    d_agent[host].eigen_decompose()
    d_agent[host].process_first_small_agent()
    d_agent[host].set_benchmark_array()
    d_agent[host].set_strand_array()
```

### Part 5: Four systems eigenvalues plot

```python
fig, ax = plt.subplots(figsize=(8,4))

for host in hosts:
    eigenvalues = d_agent[host].w
    xarray = np.array(range(len(eigenvalues)))
    label = Abbreviation.get_abbreviation(host)
    ax.plot(xarray, eigenvalues, label=label)
ax.legend()
ax.set_xlabel("Mode Index, i", fontsize=14)
ax.set_ylabel("$\lambda_i$", fontsize=14)
ax.set_title("Mean-mode eigenvalues for backbone", fontsize=16)

plt.tight_layout()
plt.savefig("/home/yizaochen/Desktop/drawzone_temp/mean_mode_eigenvalues_backbone.svg")
plt.show()
```

### Part 6: Four systems eigenvalues plot, Splot Strand

```python
hosts = ['a_tract_21mer', 'tat_21mer', 'g_tract_21mer', 'gcgc_21mer']
strand_ids = ['STRAND1', 'STRAND2']
d_results = dict()

for host in hosts:
    d_results[host] = dict()
    for strand_id in strand_ids:
        d_results[host][strand_id] = d_agent[host].get_lambda_by_strand(strand_id)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12,8))

ylims = (-0.5, 500)
ylines = range(0,501,100)

host = 'a_tract_21mer'
ax = axes[0,0]
for strand_id in strand_ids:
    eigenvalues = d_results[host][strand_id]
    xarray = np.array(range(len(eigenvalues)))
    ax.plot(xarray, eigenvalues, label=strand_id)
for yvalues in ylines:
    ax.axhline(yvalues, color="grey", alpha=0.15)
ax.legend()
title = Abbreviation.get_abbreviation(host)
ax.set_title(title,fontsize=16)
ax.set_xlabel("Mode Index, i", fontsize=14)
ax.set_ylabel("$\lambda_i$", fontsize=14)
ax.set_ylim(ylims)

host = 'tat_21mer'
ax = axes[0,1]
for strand_id in strand_ids:
    eigenvalues = d_results[host][strand_id]
    xarray = np.array(range(len(eigenvalues)))
    ax.plot(xarray, eigenvalues, label=strand_id)
for yvalues in ylines:
    ax.axhline(yvalues, color="grey", alpha=0.15)
ax.legend()
title = Abbreviation.get_abbreviation(host)
ax.set_title(title,fontsize=16)
ax.set_xlabel("Mode Index, i", fontsize=14)
ax.set_ylabel("$\lambda_i$", fontsize=14)
ax.set_ylim(ylims)

host = 'g_tract_21mer'
ax = axes[1,0]
for strand_id in strand_ids:
    eigenvalues = d_results[host][strand_id]
    xarray = np.array(range(len(eigenvalues)))
    ax.plot(xarray, eigenvalues, label=strand_id)
for yvalues in ylines:
    ax.axhline(yvalues, color="grey", alpha=0.15)
ax.legend()
title = Abbreviation.get_abbreviation(host)
ax.set_title(title,fontsize=16)
ax.set_xlabel("Mode Index, i", fontsize=14)
ax.set_ylabel("$\lambda_i$", fontsize=14)
ax.set_ylim(ylims)

host = 'gcgc_21mer'
ax = axes[1,1]
for strand_id in strand_ids:
    eigenvalues = d_results[host][strand_id]
    xarray = np.array(range(len(eigenvalues)))
    ax.plot(xarray, eigenvalues, label=strand_id)
for yvalues in ylines:
    ax.axhline(yvalues, color="grey", alpha=0.15)
ax.legend()
title = Abbreviation.get_abbreviation(host)
ax.set_title(title,fontsize=16)
ax.set_xlabel("Mode Index, i", fontsize=14)
ax.set_ylabel("$\lambda_i$", fontsize=14)
ax.set_ylim(ylims)

plt.tight_layout()
#plt.savefig("/home/yizaochen/Desktop/drawzone_temp/mean_mode_eigenvalues_split_backbone.svg")
plt.show()
```
