````python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
from networkx.drawing.nx_pylab import draw_networkx
from enmspring.networkx_display import THY_Base, CYT_Base, ADE_Base, GUA_Base, THY_Right_Base, ADE_Right_Base, CYT_Right_Base, GUA_Right_Base, dsDNA
````

### Part 1: Pentagon

````python
radius = 1.
d_nodes = dict()
name_list = ['A', 'B', 'C', 'D', 'E']
edge_list = [('A', 'B'), ('B', 'C'), ('C', 'D'), ('D', 'E'), ('E', 'A')]
angle = 18 # degree
incr_angle = 72
for name in name_list:
    angle_radian = np.deg2rad(angle)
    x = radius * np.cos(angle_radian)
    y = radius * np.sin(angle_radian)
    d_nodes[name] = (x, y)
    angle += incr_angle
position_list = [d_nodes[name] for name in name_list]
````

````python
G = nx.Graph()
G.add_nodes_from(name_list)
G.add_edges_from(edge_list)
````

````python
fig, ax = plt.subplots(figsize=(4,4))
draw_networkx(G, pos=d_nodes)
plt.show()
````
